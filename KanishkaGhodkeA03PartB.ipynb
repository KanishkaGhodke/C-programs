Z-Score Method:
The Z-score measures how many standard deviations a data point is away from the mean. Typically, data points with Z-scores beyond a certain threshold (e.g., |Z| > 3) are considered outliers.

```python
import numpy as np
from scipy import stats
z_scores = stats.zscore(df['fatal'])
abs_z_scores = np.abs(z_scores)
outliers = (abs_z_scores > 3)
outliers
```

IQR (Interquartile Range) Method:
The IQR method defines outliers as data points that fall below the first quartile minus 1.5 times the IQR or above the third quartile plus 1.5 times the IQR.

```python
Q1 = df['fatal'].quantile(0.25)
Q3 = df['fatal'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = ((df['fatal'] < lower_bound) | (df['fatal'] > upper_bound))
outliers
```

Modified Z-Score:
The modified Z-score is a variation of the Z-score that is more robust to outliers. It's calculated using the median and median absolute deviation (MAD) instead of the mean and standard deviation.

```python
median = df['fatal'].median()
mad = np.median(np.abs(df['fatal'] - median))
modified_z_scores = 0.6745 * (df['fatal'] - median) / mad
outliers = (np.abs(modified_z_scores) > 3.5)
outliers
```

We load the dataset and perform basic data tidying steps, such as removing leading/trailing whitespaces in column names, handling missing values, and converting the 'Date' column to datetime format for easy manipulation.

We define a function remove_outliers_iqr to remove outliers based on the Interquartile Range (IQR) method. It calculates the lower and upper bounds and retains only the data points within these bounds.

We apply outlier removal to selected numerical columns like 'Order_Quantity', 'Cost', 'Selling_Price', and 'Profit' using the defined function.

```python
# Convert date column to datetime format
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
df1
```

```python
# Remove leading and trailing spaces in column names
df1.columns =df1.columns.str.strip()
```

```python
numerical_columns = ['Order_Quantity', 'Cost', 'Selling_Price', 'Profit']

for col in numerical_columns:
    df1[col] = pd.to_numeric(df1[col].astype(str).str.replace('[\$,]', '', regex=True), errors='coerce')
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Apply outlier removal to numerical columns
for col in numerical_columns:
    df1 = remove_outliers_iqr(df1, col)

```
